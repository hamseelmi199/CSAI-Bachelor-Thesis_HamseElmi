```python
#Pre-requisite is creating folder with 12 empty folders and all the participants' data
#Loops through the participants file which includes the behavioural data and places a pos
#or neg label along with the file itself into the according mini-NYCq question folder

import os
import shutil
import pandas as pd
import numpy as np

# Load the CSV file into a DataFrame
csv_file_path = "filtered_file.csv"  # Update with your actual CSV file path

# Read the CSV file, assuming comma-separated values
df = pd.read_csv(csv_file_path, delimiter=',')

# Strip any leading/trailing whitespace from the column names
df.columns = df.columns.str.strip()

# Print column names to verify
print(df.columns)

# Directory containing the .set files
music_directory = 'E:\\Data Sorted\\files' # Update with your actual music directory path

# Base directory where files will be sorted
sorted_base_directory = 'E:\\Data Sorted\\Music'

# Map item numbers to their respective columns for each Music session
item_to_column = {
    1: ["Mini_Item1_Music1", "Mini_Item1_Music2", "Mini_Item1_Music3"],
    2: ["Mini_Item2_Music1", "Mini_Item2_Music2", "Mini_Item2_Music3"],
    3: ["Mini_Item3_Music1", "Mini_Item3_Music2", "Mini_Item3_Music3"],
    4: ["Mini_Item4_Music1", "Mini_Item4_Music2", "Mini_Item4_Music3"],
    5: ["Mini_Item5_Music1", "Mini_Item5_Music2", "Mini_Item5_Music3"],
    6: ["Mini_Item6_Music1", "Mini_Item6_Music2", "Mini_Item6_Music3"],
    7: ["Mini_Item7_Music1", "Mini_Item7_Music2", "Mini_Item7_Music3"],
    8: ["Mini_Item8_Music1", "Mini_Item8_Music2", "Mini_Item8_Music3"],
    9: ["Mini_Item9_Music1", "Mini_Item9_Music2", "Mini_Item9_Music3"],
    10: ["Mini_Item10_Music1", "Mini_Item10_Music2", "Mini_Item10_Music3"],
    11: ["Mini_Item11_Music1", "Mini_Item11_Music2", "Mini_Item11_Music3"],
    12: ["Mini_Item12_Music1", "Mini_Item12_Music2", "Mini_Item12_Music3"],
}

# Iterate through the .set files in the directory
for root, dirs, files in os.walk(music_directory):
    for file in files:
        if file.endswith("_Mu.set"):
            # Example file: sub01_01_Mu.set
            parts = file.split('_')
            subject_with_hyphen = parts[0]  # e.g., sub01
            subject = subject_with_hyphen.replace("-", "")  # remove the hyphen
            session = parts[1]  # e.g., 01

            # Determine the session index
            session_index = int(session) - 1  # Convert session to zero-based index

            # Print the subject and session
            print(f"Processing {file} - Subject: {subject} - Session: {session}")

            # Find the participant row
            participant_row = df[df['participant_id'] == subject]

            if not participant_row.empty:
                # Iterate over each item and move the file to the corresponding directory
                for item_number, columns in item_to_column.items():
                    if session_index < len(columns):
                        column_name = columns[session_index]
                        if column_name in participant_row.columns:
                            value = participant_row[column_name].values[0]

                            if not pd.isna(value) and value.strip():  # Check if value is not NaN and not empty string
                                # Convert value to integer
                                try:
                                    value_int = int(value)
                                except ValueError:
                                    print(f"Invalid value '{value}' for {column_name} in {file}. Skipping...")
                                    continue

                                # Determine the label based on the value
                                if 0 <= value_int <= 5:
                                    label = "neg"
                                elif 6 <= value_int <= 10:
                                    label = "pos"
                                else:
                                    print(f"Value {value_int} for {column_name} in {file} is outside the range 1-10. Skipping...")
                                    continue

                                # Determine the destination directory based on the item number
                                destination_directory = os.path.join(sorted_base_directory, str(item_number))

                                # Ensure the destination directory exists
                                os.makedirs(destination_directory, exist_ok=True)

                                # Full path to the destination file
                                new_file_name = f"{subject_with_hyphen}_{session}_{label}_Mu.set"
                                destination_file = os.path.join(destination_directory, new_file_name)
                                
                                # Copy the file to the destination directory with the new name
                                if not os.path.exists(destination_file):
                                    shutil.copy(os.path.join(root, file), destination_file)
                                    print(f"Copied {file} to {destination_file}")
                                else:
                                    print(f"File {file} already exists in {destination_file}. Skipping...")

            else:
                print(f"Participant {subject} not found in CSV file.")

            print()  # Newline for better readability
```
